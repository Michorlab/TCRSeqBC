```r

################
alpha_sample_anno<- read.table(file = "sample_anno_AlphaChain.tsv", header = TRUE, row.names = 1, sep = "\t")
tpm<- read.csv("_TPM.csv", header = TRUE, row.names = 1)
log2tpm<- log2(t(tpm) + 1)
tmp = prcomp(log2tpm)
df_pca = as.data.frame(tmp$x[,1:2], stringsAsFactors = FALSE)
table(rownames(alpha_sample_anno) == rownames(df_pca))
df_pca_anno<- cbind(df_pca, alpha_sample_anno)
df_pca_anno$Age_bin=factor(df_pca_anno$Age_bin, levels = c("young","old"))

#nrow(alpha_sample_anno)
alpha_sample_anno$Age_stage= paste(alpha_sample_anno$Stage,alpha_sample_anno$Age_bin)
nrow(alpha_sample_anno[alpha_sample_anno$Age_stage=="4 young",])

#########Heatmap of top 500 variable genes###########
log2tpm_t=t(log2tpm)
log2tpm_t=as.matrix(log2tpm_t)
sd=apply(log2tpm_t, 1, sd)
sd=as.data.frame(sd)
colnames(sd)[1]="SD"
log2tpm_t_SD=merge(log2tpm_t,sd, by=0)
rownames(log2tpm_t_SD)=log2tpm_t_SD$Row.names
log2tpm_t_SD <- log2tpm_t_SD[-which(colnames(log2tpm_t_SD) == "Row.names")]
log2tpm_t_SD1 <- log2tpm_t_SD[order(log2tpm_t_SD[,"SD"], decreasing = TRUE),]
nrow(log2tpm_t_SD1)*0.1
log2tpm_t_SD2=log2tpm_t_SD1[1:500,] #Top 10% 1837
log2tpm_t_SD3 <- log2tpm_t_SD2[-which(colnames(log2tpm_t_SD2) == "SD")]
log2tpm_t_SD4=as.matrix(log2tpm_t_SD3)
HM_mat_z=t(scale(t(log2tpm_t_SD4)))
#as.data.frame(HM_mat_z)

#alpha_sample_anno

alpha_sample_anno$Age_bin=factor(alpha_sample_anno$Age_bin,levels=c("young","old"))
alpha_sample_anno$Stage=factor(alpha_sample_anno$Stage,levels=c("0","4"))
alpha_sample_anno$Chao1_bin=factor(alpha_sample_anno$Chao1_bin,levels=c("high","low"))

colours <- list('Age_bin' = c("young" = "#D23F89","old"="#7571AF"),'Stage' = c("0" = "#009E73","4"="#E69F00"),
                'Chao1_bin' = c("high" = "#D55E00","low"="#0072B2")
                )

```

```r
#########Stat test of PC1 and PC2 ###############
df_pca_anno$Chao1_bin=factor(df_pca_anno$Chao1_bin,levels=c("high","low"))
df_pca_anno$Age_bin=factor(df_pca_anno$Age_bin,levels=c("young","old"))
df_pca_anno$CPC_Stage=factor(df_pca_anno$CPC_Stage,levels=c("0","4"))

stat.test_PC1 <- df_pca_anno %>%
  t_test(PC1 ~ Chao1_bin, var.equal=FALSE) %>%
  add_significance()%>%
  add_y_position()

stat.test_PC2 <- df_pca_anno %>%
  t_test(PC2 ~ Chao1_bin, var.equal=FALSE) %>%
  add_significance()%>%
  add_xy_position()

stat.test2_PC1 <- df_pca_anno %>%
  t_test(PC1 ~ Age_bin, var.equal=FALSE) %>%
  add_significance()%>%
  add_y_position()

stat.test2_PC2 <- df_pca_anno %>%
  t_test(PC2 ~ Age_bin, var.equal=FALSE) %>%
  add_significance()%>%
  add_xy_position()

stat.test3_PC1 <- df_pca_anno %>%
  mutate(Stage= as.factor(Stage)) %>% 
  t_test(PC1 ~ Stage, var.equal=FALSE) %>%
  add_significance()%>%
  add_xy_position()

stat.test3_PC2 <- df_pca_anno %>%
  mutate(Stage= as.factor(Stage)) %>% 
  t_test(PC2 ~ Stage, var.equal=FALSE) %>%
  add_significance()%>%
  add_xy_position()


a=ggplot(df_pca_anno, aes(x=Chao1_bin, y=PC1)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = Chao1_bin))+
  scale_colour_manual(values = c("#D55E00", "#0072B2"))+
  scale_y_continuous(limits=c(-180, 100))+
  ylab("PC1")+
  theme_pubr()+ coord_flip()

b=ggplot(df_pca_anno, aes(x=Chao1_bin, y=PC2)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = Chao1_bin))+
  scale_colour_manual(values = c("#D55E00", "#0072B2"))+
  scale_y_continuous(limits=c(-100, 40))+
  ylab("PC2")+
  xlab("Chao1")+
  theme_pubr()

c=ggplot(df_pca_anno, aes(x=Age_bin, y=PC1)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = Age_bin))+
  scale_colour_manual(values = c("#D23F89", "#7571AF"))+
  ylab("PC1")+
  xlab("Age")+
  theme_pubr()+ coord_flip()

d=ggplot(df_pca_anno, aes(x=Age_bin, y=PC2)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = Age_bin))+
  scale_colour_manual(values = c("#D23F89", "#7571AF"))+
  ylab("PC2")+
  xlab("Age")+
  theme_pubr()

e=ggplot(df_pca_anno, aes(x=CPC_Stage, y=PC1)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = CPC_Stage))+
  scale_colour_manual(values = c("#009E73", "#E69F00"))+
  ylab("PC1")+
  xlab("Stage")+
  theme_pubr()+ coord_flip()

f=ggplot(df_pca_anno, aes(x=CPC_Stage, y=PC2)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(size= 1,position = position_jitter(0.1),aes(colour = CPC_Stage))+
  scale_colour_manual(values = c("#009E73", "#E69F00"))+
  ylab("PC2")+
  xlab("Stage")+
  theme_pubr()


plot_grid(a+ clean_theme()+rremove("legend"),
          NULL,p1,b+ clean_theme()+rremove("legend"),
          ncol = 2, align = "hv",rel_widths = c(4, 1), rel_heights = c(1, 4.5))

#ggsave("001_PCA/A_Chao1.pdf", width = 6, height = 5.5)

plot_grid(c+ clean_theme()+rremove("legend"),
          NULL,p2,d+ clean_theme()+rremove("legend"),
          ncol = 2, align = "hv",rel_widths = c(4, 1), rel_heights = c(1, 4.5))

#ggsave("001_PCA/A_Age.pdf", width = 6, height = 5.5)

plot_grid(e+ clean_theme()+rremove("legend"),
          NULL,p3,f+ clean_theme()+rremove("legend"),
          ncol = 2, align = "hv",rel_widths = c(4, 1), rel_heights = c(1, 4.5))

#ggsave("001_PCA/A_Stage.pdf", width = 6, height = 5.5)
```
