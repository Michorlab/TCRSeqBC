# Sup Fig 5C

```r
###DEseq analysis#####
All_cnts=read.csv("_rawcount.csv",header=T, row.names = 1)
All_mat=as.matrix(All_cnts)
storage.mode(All_mat)="integer"
Key1=read.csv("Key.csv",header=T)
###DEseq for Chao1 high vs Chao1 low#####
Key1$Chao1=as.factor(Key1$Chao1)
Key1$Chao1 <- relevel(Key1$Chao1, ref = "low")

dds <- DESeqDataSetFromMatrix(
  countData = All_mat,
  colData = Key1,
  design= ~ Chao1)
dds <- DESeq(dds, betaPrior = FALSE)
#resultsNames(dds)

BrM2_CSF <- as.data.frame(results(dds, name="Chao1_high_vs_low"))
BrM2_CSF <- na.omit(BrM2_CSF)
BrM2_CSF_Sig=subset(BrM2_CSF,padj<0.05)
#write.csv(BrM2_CSF,"TumorRNAseq_Chao1_DEseq_output.csv")
#write.csv(BrM2_CSF_Sig,"TumorRNAseq_Chao1_DEG.csv")


########GSEA analysis#######
Key1$Chao1=as.factor(Key1$Chao1)
Key1$Chao1 <- relevel(Key1$Chao1, ref = "low")

dds <- DESeqDataSetFromMatrix(
  countData = All_mat,
  colData = Key1,
  design= ~ Chao1)

keep <- rowSums(counts(dds)) >= 10
dds <- dds[keep,]
#rm(keep)
# data tranfromation
vsd <- vst(dds, blind=FALSE)
dds <- DESeq(dds, betaPrior = FALSE)
res <- results(dds, alpha=0.05, name="Chao1_high_vs_low")
#summary(res)
res$row <- rownames(res)
ens2symbol <- AnnotationDbi::select(org.Hs.eg.db,
                                    key=res$row, 
                                    columns="ENSEMBL",
                                    keytype="SYMBOL")

names(ens2symbol)[1] <- "row"
ens2symbol <- as_tibble(ens2symbol)
#ens2symbol
res <- merge(data.frame(res), ens2symbol, by=c("row"))
res2 <- res %>% 
  ###select(row, stat) %>% 
  na.omit() %>% 
  distinct() %>% 
  group_by(row) %>% 
  summarize(stat=mean(stat))

ranks <- res2$stat
names(ranks) <- res2$row


#####hallmark_all#####
pathways.c2.cp <- gmtPathways("../997_Ontology/h.all.v7.5.1.symbols.gmt")
#head(pathways.c2.cp)
fgseaRes <- fgseaMultilevel(pathways=pathways.c2.cp, stats=ranks)
fgseaResTidy <- fgseaRes %>%
  as_tibble() %>%
  arrange(desc(NES)) 
gene.in.pathway <- pathways.c2.cp %>% 
  enframe("pathway", "row") %>% 
  unnest(cols = c(row)) %>% 
  inner_join(res, by="row")
fgseaResTidy$adjPvalue <- ifelse(fgseaResTidy$padj <= 0.05, "significant", "non-significant")
#fwrite(fgseaResTidy, file="GSEA_hallmark_Chao1_high_vs_low_all3.txt", sep="\t", sep2=c("", " ", ""))
test <- data.frame(fgseaResTidy)
test_sig = test[test$adjPvalue == "significant",]
test_sig_pos = test_sig[test_sig$NES>=0,]
test_2 = test_sig[order(test_sig$NES),]
test_sig_neg = test_2[test_2$NES<=0,]
####test_pos = test_sig[0:5,]
#test_sig_pos

########
test_sig_pos$GSEA="Hallmark_UP"
test_sig_neg$NES=-test_sig_neg$NES
test_sig_neg$GSEA="Hallmark_DOWN"
test_sig_X=rbind(test_sig_pos,test_sig_neg)
test_sig_X$GSEA=factor(test_sig_X$GSEA,level=c("Hallmark_UP","Hallmark_DOWN"))
test_sig_X %>%
  mutate(pathway = fct_reorder(pathway, -log10(padj))) %>%
  ggplot(aes(x=-log10(padj), y=pathway, size=NES,col=GSEA))+
  scale_color_manual(values = c("#D55E00", "#0072B2"))+
  geom_point() +
  facet_grid(GSEA ~ ., scales = "free", space = "free")+
  theme_pubr()



p1=test_sig_pos %>%
  mutate(pathway = fct_reorder(pathway, -log10(padj))) %>%
  ggplot(aes(x=-log10(padj), 
             y=pathway, 
             size=NES)) +
  geom_point(color = "#D55E00") +
  expand_limits(x=0) +
  theme_pubr(legend = c("right")) +
  theme_cleveland()+
  labs(x="-log10(FDR)", y="Enriched in Chao1-high", size="NES")+
  scale_x_continuous(limits = c(0, 15))

p2 = test_sig_neg %>%
  mutate(pathway = fct_reorder(pathway, -log10(padj))) %>%
  ggplot(aes(x=-log10(padj), 
             y=pathway, 
             size=-NES)) +
  geom_point(color = "#0072B2") +
  expand_limits(x=0) +
  theme_pubr(legend = c("right")) +
  theme_cleveland()+
  labs(x="-log10(FDR)", y="Enriched in Chao1-low", size="NES")+
  scale_x_continuous(limits = c(0, 5))

#plot_grid(p1, p2, ncol = 2, align = "v")
```
