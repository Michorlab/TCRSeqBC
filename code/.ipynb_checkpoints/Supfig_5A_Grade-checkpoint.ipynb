```r
# Load required libraries
library(ggplot2)
library(ggsignif)
library(dplyr)

# Load your data
Alpha <- read.csv("Alpha_diverisity_489_2.csv", header = TRUE, row.names = 1)
Alpha2 <- Alpha[c(3, 15, 16, 45, 53)]
Alpha3 <- Alpha2[Alpha2$Stage == "0", ]
Alpha4 <- na.omit(Alpha3)
write.csv(Alpha4, "Grade_Chao1.csv")
Alpha5 <- Alpha4[Alpha4$Timepoint == "Baseline", ]
```

```r
# Perform a t-test between groups
stat.test <- compare_means(
 Chao1 ~ Grade, data = Alpha5,
 method = "wilcox.test"
)
stat.test
```

```r
Alpha5$Chao1_bin=factor(Alpha5$Chao1_bin,levels=c("high","low","#N/A"))
p<-ggplot(Alpha5, aes(x=Grade, y=Chao1)) +
  geom_boxplot(outlier.shape = NA)+
  geom_jitter(position = position_jitter(0.1),aes(colour = Chao1_bin))+
  scale_colour_manual(values = c("#D55E00", "#0072B2", "black"))+
  ylab("Chao1")+
  xlab("Grade")+
  scale_x_discrete(limit = c("I", "II", "III"),labels = c("Grade 1","Grade 2","Grade 3"))+
  #stat_pvalue_manual(stat.test3, label = "p.adj")+
  theme_pubr(legend=c("right"))

stat.test <- stat.test %>%
 mutate(y.position = c(40000, 45000, 50000))
p + stat_pvalue_manual(stat.test, label = "p.adj")
```
