```r
result <- merge(x=cdr3_data2,y=meta,by="sampleID",all.y=TRUE)
names(result)[2] <- "aminoAcid"


clonotype_freq<- result  %>% dplyr::group_by(sampleID) %>% 
  dplyr::mutate(percentage = count_sum/sum(count_sum)*100)
clonotype_freq$Age_Stage <- paste(clonotype_freq$Age_bin,clonotype_freq$Stage) 
#clonotype_freq

data_freq_bin <- clonotype_freq %>% mutate(percent = case_when(
  percentage < 0.01                   ~ "Rare (0<x<1e-4)",
  percentage >= 0.01 & percentage <1  ~ "Medium (1e-4<=x<1e-2)",
  percentage >= 1     ~ "Large (1e-2<=x<1)"
))


data_freq_bin$percent=factor(data_freq_bin$percent,levels = c("Rare (0<x<1e-4)","Medium (1e-4<=x<1e-2)","Large (1e-2<=x<1)"))
X=aggregate(x=data_freq_bin[c("percentage")], by=list(data_freq_bin$sampleID,data_freq_bin$percent,data_freq_bin$Age_Stage), sum)
X1=pivot_wider(X, names_from = Group.2, values_from = percentage, values_fill = 0)

X1$order=X1[,4]+X1[,5]
data_freq_bin5=arrange(X1, desc(order))
names(data_freq_bin5)[1] <- "sampleID"
names(data_freq_bin5)[2] <- "Age_Stage"
names(data_freq_bin5)[8] <- "order"

data_freq_bin5$sampleID=factor(data_freq_bin5$sampleID,levels = data_freq_bin5$sampleID)
X2=data_freq_bin5  %>% pivot_longer(cols = c('Rare (0<x<1e-4)':'Large (1e-2<=x<1)'), names_to = "percent", values_to = "percentage")
X2$percent=factor(X2$percent,levels = c("Rare (0<x<1e-4)","Medium (1e-4<=x<1e-2)","Large (1e-2<=x<1)"))

# create the plot


#####stack########
#data_freq_bin["percentage"][is.na(data_freq_bin["percentage"])] <- 0
data_old4 <- X2[X2$Age_Stage == "old 4",]
data_old0 <- X2[X2$Age_Stage == "old 0",]
data_young0 <- X2[X2$Age_Stage == "young 0",]
data_young4 <- X2[X2$Age_Stage == "young 4",]


```

```r
library(tidyverse)
library(ggplot2)
library(ggh4x)
library(dplyr)
library(ggpubr)
library(rstatix)


giana<- read.table("data/figure2/cdr3_alpha_giana_output_06_23_2023.txt", header = FALSE, sep = "\t")
colnames(giana) <- c("CDR3", "cluster", "sampleID", "count", "frequency", "Stage", "Age", "Age_bin", "Chao1")


# Count the number of different cluster categories for each patient
patient_counts <- giana %>%
  group_by(sampleID) %>%
  summarise(unique_clusters = n_distinct(cluster))

Pat_info=unique(select(giana, sampleID, Stage, Age_bin))
patient_counts=as.data.frame(patient_counts)
patient_counts2<- merge(x=patient_counts, y=Pat_info, by="sampleID")

# Create the box plot with dots
patient_counts2$Age_stage <- paste(patient_counts2$Age_bin, patient_counts2$Stage)
stat.test <- aov(unique_clusters ~ Age_stage, data = patient_counts2) %>%
  tukey_hsd()%>%
  add_y_position(formula= unique_clusters ~ Age_stage, data = patient_counts2,step.increase = 0.3)

#write.csv(patient_counts2,"GIANA_data.csv")

patient_counts2 %>%mutate(Age_stage = factor(patient_counts2$Age_stage, levels = c("young 0","young 4", "old 0","old 4")))  %>%
  ggplot(aes(x = Age_stage, y = unique_clusters, color=Age_stage)) +
  geom_boxplot(outlier.shape = NA, color = "Black", ) +
  #geom_text(aes(label=sprintf("%0.2f", `Pr(>|z|)`))
  labs(y = "Number of clusters") +
  geom_jitter(aes(fill= Age_stage, color= Age_stage),size = 0.7, width = 0.25, alpha = 0.6) +
  scale_fill_manual(values = c("#0073C2FF", "#EFC000FF", "#868686FF", "#CD534CFF")) +
  scale_color_manual(values = c("#0073C2FF", "#EFC000FF", "#868686FF", "#CD534CFF")) +
  #stat_compare_means(method="wilcox.test", label = "p.format", vjust = 0.5,  hjust = 0, bracket.size = 0.3) +
  scale_y_continuous(limits = function(y) c(0,max(y))) + 
  stat_pvalue_manual(stat.test,label = "p = {p.adj}")+
  #stat_compare_means(aes(label = ifelse(p < 1.e-2,
  #sprintf("p = %2.1e", as.numeric(..p.format..)),
  #sprintf("p = %5.4f", as.numeric(..p.format..)))))+
  theme_pubr() + 
  theme(axis.ticks.x=element_blank(), legend.position="right")+ theme(legend.position = "none")

```

```r

library(ggplot2)
library(scales)
library(ggpubr)
library(cowplot)


#### bar plot for clonotype frequency##########
meta <- read.csv("alpha_diverisity_485_b_small_capital.csv")
#Read the files
cdr3_all <- read.csv("cdr3_all_alpha.csv")

#Subsetting the meta file
meta <- meta[c( "sampleID", "Chao1", "Age_bin", "Stage")]

# Remove complete duplicates
cdr3_data<- unique(cdr3_all)

# Filter out low-quality sequences
cdr3_data <- cdr3_data[cdr3_data$cdr3 != "NA",]


cdr3_data2=aggregate(x=cdr3_data[c("count_sum")], by=list(cdr3_data$sheet,cdr3_data$cdr3), sum)

# Filter out sequences with low read counts
min_read_count <- 1
cdr3_data2 <- cdr3_data2[cdr3_data2$count_sum >= min_read_count,]
length(unique(cdr3_data2$cdr3))
#Final input file for GIANNA
colnames(cdr3_data2) [1] <- "sampleID"
colnames(cdr3_data2) [2] <- "cdr3"

cdr3_data2$Freq=1


meta$Age_Stage <- paste(meta$Age_bin,meta$Stage)
meta_data_old4 <- meta[meta$Age_Stage == "old 4",]
meta_data_old0 <- meta[meta$Age_Stage== "old 0",]
meta_data_young0 <- meta[meta$Age_Stage == "young 0",]
meta_data_young4 <- meta[meta$Age_Stage == "young 4",]
meta_data_old4

```

```r
df_list <- list(meta_data_young0, meta_data_young4, meta_data_old0, meta_data_old4)
Pick <- function(df) {
  df = merge(cdr3_data2,df, by=c("sampleID"))
  df$Total = length(unique(df$sampleID))
  df = aggregate(x=df[c("Freq")], by=list(df$cdr3,df$Total), sum)
  df <- df[order(df$Freq, decreasing = TRUE),]
  colnames(df) [1] <- "sampleID"
  colnames(df) [2] <- "Total"
  df$Percent=df$Freq/df$Total
  df$Count = 1
  df = aggregate(x=df[c("Count")], by=list(df$Freq), sum)
  colnames(df) [1] <- "NPatient_overlap"
  return(df)
}
processed_list <- lapply(df_list, Pick)


young0=as.data.frame(processed_list[1])
a=ggplot(young0,aes(x=NPatient_overlap,y=Count))+
  geom_col(fill="#0073C2FF",color="black", linewidth = 0.2)+
  scale_y_continuous(
    trans = pseudo_log_trans(base = 10),
    breaks = c(0, 10^(0:5),500000),
    #labels = label_number(accuracy = 0.01)
  )+
  ylab("Number of CDR3 sequences")+
  xlab("Number of patients")+
  theme_bw()
#ggsave("figure2/Overlap_Young0.pdf", width = 6, height = 3)

young4=as.data.frame(processed_list[2])
b=ggplot(young4,aes(x=NPatient_overlap,y=Count))+
  geom_col(fill="#E69F00",color="black", linewidth = 0.2)+
  scale_y_continuous(
    trans = pseudo_log_trans(base = 10),
    breaks = c(0, 10^(0:5),500000),
    #labels = label_number(accuracy = 0.01)
  )+
  ylab("Number of CDR3 sequences")+
  xlab("Number of patients")+
  theme_bw()
#ggsave("figure2/Overlap_Young4.pdf", width = 6, height = 3)

old0=as.data.frame(processed_list[3])
c=ggplot(old0,aes(x=NPatient_overlap,y=Count))+
  geom_col(fill="#868686FF",color="black", linewidth = 0.2)+
  scale_y_continuous(
    trans = pseudo_log_trans(base = 10),labels = function(x) format(x, scientific = FALSE),
    breaks = c(0, 10^(0:5),500000),
    #labels = label_number(accuracy = 0.01)
  )+
  ylab("Number of CDR3 sequences")+
  xlab("Number of patients")+
  theme_bw()
#ggsave("figure2/Overlap_Old0.pdf", width = 6, height = 3)

old4=as.data.frame(processed_list[4])
d=ggplot(old4,aes(x=NPatient_overlap,y=Count))+
  geom_col(fill="#CD534CFF",color="black", linewidth = 0.2)+
  scale_y_continuous(
    trans = pseudo_log_trans(base = 10),
    breaks = c(0, 10^(0:5), 500000),
    #labels = label_number(accuracy = 0.01)
  )+
  ylab("Number of CDR3 sequences")+
  xlab("Number of patients")+
  theme_bw()
#ggsave("figure2/Overlap_Old4.pdf", width = 6, height = 3)

z=plot_grid(a, b,c,d, ncol = 2, align = "v")
#ggsave("figure2/Overlap.pdf", width = 10, height = 6)

```

```r
library(tidyverse)
library(ggplot2)
library(ggh4x)
library(dplyr)
library(ggpubr)
library(rstatix)

#Unique
Overall=c(126,871,125,266)
names(Overall)=c("Young_yes","Young_no","Old_yes","Old_no")

Young_yes <- read.csv("Count1/Young_DCIS_recurrence_unique/McPAS-TCR_search_recurrence_yes.csv")
Young_no <- read.csv("Count1/Young_DCIS_recurrence_unique/McPAS-TCR_search_recurrence_no.csv")
Old_yes <- read.csv("Count1/Old_DCIS_recurrence_unique/McPAS-TCR_search_yes.csv")
Old_no <- read.csv("Count1/Old_DCIS_recurrence_unique/McPAS-TCR_search_no.csv")

Identified=list(length(unique(Young_yes$Number.in.Query)),length(unique(Young_no$Number.in.Query)),length(unique(Old_yes$Number.in.Query)),length(unique(Old_no$Number.in.Query)))
Identified=as.data.frame(Identified)
names(Identified)=c("Young_yes","Young_no","Old_yes","Old_no")
Unidentified=Overall-Identified
#Unidentified

Recurrence=list(Young_yes,Young_no,Old_yes,Old_no)

Categorize <- function(df){
  df$count=1
  df=aggregate(x=df[c("count")], by=list(df$Number.in.Query,df$Category), sum)
  colnames(df)=c("Cluster","Category","Sum")
  df=df[order(df$Cluster,decreasing=F),]
  df <- df %>%
  group_by(Cluster) %>%
  mutate(Category_new = ifelse(n() > 1, paste(Category, collapse = "/"), Category)) %>%
  ungroup()
  df=unique(select(df,Cluster,Category_new))
  df$count=1
  df=aggregate(x=df[c("count")], by=list(df$Category_new), sum)
  colnames(df)=c("Category","Sum")
  return(df)
}
processed_list <- lapply(Recurrence, Categorize)
#processed_list
#Unidentified

for (i in 1:length(processed_list)) {
  new_row <- data.frame(A = "Unidentified", B = Unidentified[1, i])
  colnames(new_row) <- colnames(processed_list[[i]])
  processed_list[[i]] <- rbind(processed_list[[i]], new_row)
  processed_list[[i]]$Percentage=processed_list[[i]]$Sum/sum(processed_list[[i]]$Sum)*100
}
names(processed_list)=c("Young_yes","Young_no","Old_yes","Old_no")
#processed_list
df <- processed_list %>%
  bind_rows(.id = "Group")
colors <- scales::hue_pal()(14)
df$Group=factor(df$Group,levels=c("Young_yes","Young_no","Old_yes","Old_no"))

bar<-ggplot(df, aes(x = Group, y = Percentage, fill=Category)) +
  geom_bar(stat = "identity",width = 0.75) +
  xlab("Group") +
  ylab("Neoantigen type") +
  ggtitle("Neoantigen") +
  #facet_wrap(~Stage,  ncol = 1)
  #facet_grid(~Stage, scale='free_x')+
  #force_panelsizes(cols = c(0.5, 1)) +
  #scale_fill_discrete(labels = c(Stage4 & old"))
  theme_pubr(legend=c("right"))+
  #theme(axis.text.x = element_blank(),axis.ticks.x = element_blank())+
  scale_fill_manual(values=colors)
  #scale_fill_brewer(palette="Spectral", direction=-1)

#ggsave("Neoantigen_type_all.pdf", width = 8, height=8)

######Summerize cancer-related antigens#######
processed_list <- lapply(Recurrence, Categorize)
processed_list2=processed_list

for (i in 1:length(processed_list2)) {
  processed_list2[[i]]$Category_cancer <- ifelse(grepl("Cancer", processed_list2[[i]]$Category), "Cancer", "Allergy/Autoimmune/Pathogens/Others")
  processed_list2[[i]]=aggregate(x=processed_list2[[i]][c("Sum")], by=list(processed_list2[[i]]$Category_cancer), sum)
  colnames(processed_list2[[i]])=c("Category","Sum")
}
#processed_list2

for (i in 1:length(processed_list2)) {
  new_row <- data.frame(A = "Unidentified", B = Unidentified[1, i])
  colnames(new_row) <- colnames(processed_list2[[i]])
  processed_list2[[i]] <- rbind(processed_list2[[i]], new_row)
  processed_list2[[i]]$Percentage=processed_list2[[i]]$Sum/sum(processed_list2[[i]]$Sum)*100
}
names(processed_list2)=c("Young_yes","Young_no","Old_yes","Old_no")
#processed_list2
df <- processed_list2 %>%
  bind_rows(.id = "Group")
colors <- scales::hue_pal()(3)

df$Group=factor(df$Group,levels=c("Young_yes","Young_no","Old_yes","Old_no"))
df$Category=factor(df$Category,levels=c("Unidentified","Allergy/Autoimmune/Pathogens/Others","Cancer"))
ggplot(df, aes(x = Group, y = Percentage, fill=Category)) +
  geom_bar(stat = "identity",width = 0.75) +
  xlab("Unique GIANA cluster in each group w/o recurrence") +
  ylab("Percentage of CDR3 clusters") +
  ggtitle("Category of predicted antigens") +
  #facet_wrap(~Stage,  ncol = 1)
  #facet_grid(~Stage, scale='free_x')+
  #force_panelsizes(cols = c(0.5, 1)) +
  #scale_fill_discrete(labels = c(Stage4 & old"))
  theme_pubr(legend=c("right"))+
  #theme(axis.text.x = element_blank(),axis.ticks.x = element_blank())+
  scale_fill_manual(values=c("grey","#619CFF","#F8766D"))
#scale_fill_brewer(palette="Spectral", direction=-1)

#ggsave("Neoantigen_type_cancer_others2.pdf", width = 8, height=6)

df1=df[,-4]

X1=as.data.frame(pivot_wider(df1, names_from = Group, values_from = Sum, values_fill = 0))
rownames(X1)=X1$Category

Chi_young=X1[2:3]
#Chi_young
chisq.test(Chi_young)

#df
#15/(31+15+80)

data <- data.frame(
  Young_yes = c(15, 31+80),
  Young_no = c(157, 296+418)
)
rownames(data)=c("Cancer","Others")
#chisq.test(data,correct = F)


Chi_old=X1[4:5]
#Chi_old
chisq.test(Chi_old)



```
