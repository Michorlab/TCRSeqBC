# Alpha wilcoxon test and fdr adjustment

```r
# Creating a list of indices
index_list <- c("Shannon", "Clonality", "Richness", "Chao1", "Simpson")

# Create an empty data frame to store the results
results_df <- data.frame(
  Index = character(),
  Group_Comparison = character(),
  P_Value = numeric(),
  Adjusted_P_Value_Index = numeric(),
  Adjusted_P_Value_All = numeric(),
  stringsAsFactors = FALSE
)

# Step 3: Perform pairwise Wilcoxon rank-sum tests and store p-values in results_df
for (index in index_list) {
  for (group1 in unique(df$Age_stage)) {
    for (group2 in unique(df$Age_stage)) {
      if (group1 != group2 && !(group1 == "young 4" && group2 == "old 0") && !(group1 == "young 0" && group2 == "old 4")
         && !(group1 == "old 0" && group2 == "young 4") && !(group1 == "old 4" && group2 == "young 0")) {
        # Subset the data for each group comparison
        data_subset <- df[df$Age_stage %in% c(group1, group2), ]
        
        # Perform pairwise Wilcoxon rank-sum test
        wilcox_result <- wilcox.test(data_subset[[index]] ~ data_subset$Age_stage)
        
        # Store the results in the data frame
        results_df <- rbind(results_df, data.frame(
          Index = index,
          Group_Comparison = paste(group1, group2, sep = "-"),
          P_Value = wilcox_result$p.value,
          Adjusted_P_Value_Index = NA,
          Adjusted_P_Value_All = NA
        ))
      }
    }
  }
  
  # Calculate FDR-adjusted p-values within each index
  results_df_subset <- results_df[results_df$Index == index, ]
  results_df_subset$Adjusted_P_Value_Index <- p.adjust(results_df_subset$P_Value, method = "fdr")
  results_df[results_df$Index == index, "Adjusted_P_Value_Index"] <- results_df_subset$Adjusted_P_Value_Index
  
  
}

# Calculate FDR-adjusted p-values across all indices
results_df$Adjusted_P_Value_All <- p.adjust(results_df$P_Value, method = "fdr")
# Round the raw p-values and adjusted p-values to four digits
results_df$P_Value <- round(results_df$P_Value, digits = 4)
results_df$Adjusted_P_Value_Index <- round(results_df$Adjusted_P_Value_Index, digits = 4)
results_df$Adjusted_P_Value_All <- round(results_df$Adjusted_P_Value_All, digits = 4)

# Split the group comparison column into two groups 
library(tidyr)
results_df <- separate(data = results_df, col = Group_Comparison, into = c("group1", "group2"), sep = "-")
colnames(results_df)[6] <- "p.adj"

# Print the results
#print(results_df)

# Write the results to a CSV file
#write.csv(results_df, "Alpha_Age_Stage_Wilcoxon_fdr_adjustment.csv", quote = FALSE)


library(dplyr)

# Assuming your data frame is named 'df'
alpha_unique_results_df <- results_df %>%
  mutate(
    sorted_groups = pmap_chr(list(group1, group2), ~ paste(sort(c(..1, ..2)), collapse = "-"))
  ) %>%
  group_by(Index, sorted_groups) %>%
  filter(row_number() == 1) %>%
  select(-sorted_groups)

#write.csv(alpha_unique_results_df, "Alpha_Age_Stage_wilcoxon_fdr_adjustment.csv", quote=FALSE)
alpha_unique_results_df


```

```r
# Assuming your data frame is named 'df'

# Filter the dataframe to remove rows where "Recurrence_bin" is NA and "Stage" is 0
recur_df <- df %>%
  filter(!is.na(Recurrence_bin)) %>% 
  filter(Stage == 0)

# Creating a list of indices
index_list <- c("Shannon", "Clonality", "Richness", "Chao1", "Simpson")

# Create an empty data frame to store the results
wilcoxon_df_old <- data.frame(
  Index = character(),
  Group_Comparison = character(),
  P_Value = numeric(),
  stringsAsFactors = FALSE
)

wilcoxon_df_young <- data.frame(
  Index = character(),
  Group_Comparison = character(),
  P_Value = numeric(),
  stringsAsFactors = FALSE
)

# Perform Wilcoxon rank-sum tests separately for "old" and "young" patients and store p-values in wilcoxon_df
for (index in index_list) {
  # Perform Wilcoxon rank-sum test for "old" patients     
  df_old <- recur_df[recur_df$Age_bin == "old", ]  
  wilcoxon_result_old <- wilcox.test(df_old[[index]] ~ as.factor(df_old$Recurrence_bin))
  
  # Store the results in the data frame for "old" patients
  wilcoxon_df_old <- rbind(wilcoxon_df_old, data.frame(
    Index = index,
    Group_Comparison = "yes-no",
    Age_Bin = "old",
    P_Value = round(wilcoxon_result_old$p.value, digits=5)
  ))
}

# Perform Wilcoxon rank-sum tests separately for "old" and "young" patients and store p-values in wilcoxon_df
for (index in index_list) {
  # Perform Wilcoxon rank-sum test for "young" patients
  df_young <- recur_df[recur_df$Age_bin == "young", ] 
  wilcoxon_result_young <- wilcox.test(df_young[[index]] ~ as.factor(df_young$Recurrence_bin))
  
  # Store the results in the data frame for "young" patients
  wilcoxon_df_young <- rbind(wilcoxon_df_young, data.frame(
    Index = index,
    Group_Comparison = "yes-no",
    Age_Bin = "young",
    P_Value = round(wilcoxon_result_young$p.value, digits=5)
  ))
}


recur_wilcoxon_df <- rbind(wilcoxon_df_old, wilcoxon_df_young)

# Calculate FDR-adjusted p-values for the combined "old" and "young" comparisons
recur_wilcoxon_df$Adjusted_P_Value <- round(p.adjust(recur_wilcoxon_df$P_Value, method = "fdr"), digits=5)

# Split the group comparison column into two groups 
library(tidyr)
recur_wilcoxon_df <- separate(data=recur_wilcoxon_df, col=Group_Comparison , into=c("group1", "group2"), sep="-")
colnames(recur_wilcoxon_df)[6] <- "p.adj"

#write.csv(wilcoxon_df, "Alpha_Recurrence_ttest_fdr_adjustment.csv", quote=FALSE)
recur_wilcoxon_df
```

```r
# Assuming your data frame is named 'df'

# Filter the dataframe to remove rows where "Recurrence_bin" is NA and "Stage" is 0
recur_df <- df %>%
  filter(!is.na(Recurrence_bin)) %>% 
  filter(Stage == 0)

# Creating a list of indices
index_list <- c("Shannon", "Clonality", "Richness", "Chao1", "Simpson")


# Create an empty data frame to store the results
recur_wilcoxon_df <- data.frame(
  Index = character(),
  Age_Bin = character(),
  Groups = character(),
  P_Value = numeric(),
  p.adj = numeric(),
  stringsAsFactors = FALSE
)

# Perform Wilcoxon rank-sum tests and store p-values in wilcoxon_df
for (index in index_list) {
  for (age_bin in unique(recur_df$Age_bin)) {
    # Perform Wilcoxon rank-sum test for each combination of index and age_bin
    df_age_bin <- recur_df[recur_df$Age_bin == age_bin, ]
    wilcox_result <- wilcox.test(df_age_bin[[index]] ~ as.factor(df_age_bin$Recurrence_bin))
    
    # Store the results in the data frame
    recur_wilcoxon_df <- rbind(recur_wilcoxon_df, data.frame(
      Index = index,
      Groups = "yes-no",
      Age_Bin = age_bin,
      P_Value = wilcox_result$p.value,
      p.adj = NA
    ))
  }
  
  # Calculate FDR-adjusted p-values within each index
  recur_wilcoxon_df_subset <- recur_wilcoxon_df[recur_wilcoxon_df$Index == index, ]
  recur_wilcoxon_df_subset$p.adj <- p.adjust(recur_wilcoxon_df_subset$P_Value, method = "fdr")
  recur_wilcoxon_df[recur_wilcoxon_df$Index == index, "p.adj"] <- recur_wilcoxon_df_subset$p.adj
}

# Calculate FDR-adjusted p-values across all indices
recur_wilcoxon_df$Adjusted_P_Value <- p.adjust(recur_wilcoxon_df$P_Value, method = "fdr")

# Split the group comparison column into two groups 
library(tidyr)
recur_wilcoxon_df <- separate(data=recur_wilcoxon_df, col=Groups , into=c("group1", "group2"), sep="-")
colnames(recur_wilcoxon_df)[6] <- "p.adj"

#write.csv(wilcoxon_df, "Alpha_Recurrence_ttest_fdr_adjustment.csv", quote=FALSE)
recur_wilcoxon_df
```

```r
########Correlation#######

df_long<- df  %>% pivot_longer(c(Chao1:D4,Simpson,param.alpha), names_to = "diversity")
#write_tsv(df_long, "Alpha_diverisity_baseline_long.tsv")
df_long$Stage<- df_long$Stage  %>% as.factor()

df_long4=df_long[df_long$diversity== "Richness"|df_long$diversity== "Chao1"|df_long$diversity=="Clonality"|df_long$diversity=="Shannon"|df_long$diversity=="Simpson",]
df_long4$diversity=factor(df_long4$diversity,levels=c("Richness","Chao1","Shannon","Clonality","Simpson"))

df_long4 %>%
  ggplot(aes(x = Age, y = value, color=as.factor(Stage))) +
  geom_point(size=0.7, alpha = 0.6) +
  labs(y = "TCR Alpha Chain Diversity") +
  scale_color_manual(values = c('#009E73', '#E69F00')) +
  scale_y_continuous(limits = function(y) c(0,max(y))) +
  facet_grid2(Stage ~ diversity, scales = "free", independent = "y") +
  theme(text = element_text(size = 10), legend.position="right")+geom_smooth(color="black",method=lm, size=0.5)+ 
                     stat_cor(color="black",method = "pearson", digits = 3)+
  theme_pubr()+ theme(legend.position = "none")

#ggsave("Scatter4.pdf", width = 10, height=5)

                     
# Compute Pearson correlation and obtain raw p-value
alpha_correlation_result <- df_long4 %>%
  group_by(Stage, diversity) %>%
  summarize(correlation = cor(Age, value, method = "pearson"),
            p_value = round(cor.test(Age, value, method = "pearson")$p.value, digits = 7),
            .groups = "keep")
                    
# Combine all p-values into a single vector
all_p_values <- alpha_correlation_result$p_value

# Adjust the combined p-values using BH method
adjusted_p_values <- p.adjust(all_p_values, method = "BH")

# Add the adjusted p-values to the alpha_correlation_result data frame
alpha_correlation_result$adj_p_value <- round(adjusted_p_values, digits = 5)

#write.csv(alpha_correlation_result, "Alpha_correlation_BH_adjustment.csv", quote=FALSE)
#head(beta_results_df)
```
